```python
import pandas as pd
```

```python
df = pd.read_csv("IRIS1.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
# 150 rows and 5 columns
#handle missing values in all rows
```

```python
df.isnull().sum()
```

```python
df['sepal_length']=df['sepal_length'].fillna(df['sepal_length'].mean())
```

```python
df.isnull().sum()
```

```python
df['sepal_width']=df['sepal_width'].fillna(df['sepal_width'].mean())
```

```python
df.isnull().sum()
```

```python
df['petal_width'] = df['petal_width'].fillna(df['petal_width'].mean())
```

```python
df.isnull().sum()
```

```python
df['petal_length'] = df['petal_length'].fillna(df['petal_length'].mean())
```

```python
df.isnull().sum()
```

```python
df['species'] = df['species'].fillna(df['species'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.to_csv("cleaned_iris.csv")
```
